```python
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
data_X, data_y = load_wine( return_X_y = True )

# pega todas as linhas, e apenas coluna 0 e 1
entrada = data_X[:]
saida = data_y[:]
```

Tentativas de escolha

Para escolher o melhor Kernel, gamma e C, busca-se um que o model.score seja o mais próximo possível de 1.
Assim, foram feitos varios testes.

```python
model = svm.SVC(kernel="rbf", degree=2, gamma='scale', C=5)
model.fit(entrada, saida)
print("Model score tentativa 1 = ", model.score(entrada, saida))

model = svm.SVC(kernel="linear", degree=5, gamma='auto', C=1)
model.fit(entrada, saida)
print("Model score tentativa 2 = ", model.score(entrada, saida))

model = svm.SVC(kernel="linear", degree=5, gamma='auto', C=10)
model.fit(entrada, saida)
print("Model score tentativa 3 = ", model.score(entrada, saida))

model = svm.SVC(kernel="poly", degree=7, gamma='auto', C=0.3)
model.fit(entrada, saida)
print("Model score tentativa 4 = ", model.score(entrada, saida))

model = svm.SVC(kernel="poly", degree=7, gamma='auto', C=5)
model.fit(entrada, saida)
print("Model score tentativa 5 = ", model.score(entrada, saida))

model = svm.SVC(kernel="sigmoid", degree=3, gamma='auto', C=0.7)
model.fit(entrada, saida)
print("Model score tentativa 6 = ", model.score(entrada, saida))

model = svm.SVC(kernel="sigmoid", degree=3, gamma='auto', C=10)
model.fit(entrada, saida)
print("Model score tentativa 7 = ", model.score(entrada, saida))
```

Após a escolha dos modelos, foi feita uma divisão no dataset entre treinamento e testes, em que 80% do modelo sirva para treinamento e 20% para testes.
Isso serve para que o modelo não fique especialista no treinamento, e ao fornecer dados novos, ele não funciona.
Dessa forma, é possível avaliar como estão os testes a partir de uma parte dos dados do dataset, verificando a eficácia do treinamento.

```python
# treinamento com um pedaco da database fornecida
from sklearn.model_selection import train_test_split

entrada_treinamento, entrada_teste, saida_treinamento, saida_teste = train_test_split(entrada, saida, test_size=0.20)

len(entrada_teste)
```

Com a divisão feita, utiliza-se o relatório de classificação presente na biblioteca para avaliar a precisão, f1-score e recall de cada classe, além de a acurácia geral do modelo. É a partir desses dados que podemos avaliar a eficácia do treinamento. Afinal, é possível que em alguns testes ele se saia muito bem, mas ao avaliar ele não realizou testes com determinada propriedade por exemplo, ou errou muito uma das classificações apesar de acertar outras. 

Assim, foi feito o relatório de classificação para os primeiros modelos, para analisar a diferença entre eles.

```python
model = svm.SVC(kernel="rbf", degree=2, gamma='scale', C=5)
model.fit(entrada_treinamento, saida_treinamento)
```

```python
from sklearn.metrics import classification_report

previsto = model.predict(entrada_teste)

print(classification_report(saida_teste, previsto))
```

```python

model = svm.SVC(kernel="linear", degree=5, gamma='auto', C=1)
model.fit(entrada_treinamento, saida_treinamento)
```

```python
from sklearn.metrics import classification_report

previsto = model.predict(entrada_teste)

print(classification_report(saida_teste, previsto))
```

```python

model = svm.SVC(kernel="linear", degree=5, gamma='auto', C=10)
model.fit(entrada_treinamento, saida_treinamento)
```

```python
from sklearn.metrics import classification_report

previsto = model.predict(entrada_teste)

print(classification_report(saida_teste, previsto))
```

```python

model = svm.SVC(kernel="poly", degree=7, gamma='auto', C=0.3)
model.fit(entrada_treinamento, saida_treinamento)
```

```python
from sklearn.metrics import classification_report

previsto = model.predict(entrada_teste)

print(classification_report(saida_teste, previsto))
```

```python

model = svm.SVC(kernel="poly", degree=7, gamma='auto', C=5)
model.fit(entrada_treinamento, saida_treinamento)
```

```python
from sklearn.metrics import classification_report

previsto = model.predict(entrada_teste)

print(classification_report(saida_teste, previsto))
```

Para o teste de eficácia dos treinamentos, é possível perceber que o modelo com Kernel rbf teve um péssimo desempenho, sendo que dentre os analisados com a classificação, foi aquele que o modelo obteve pior score.

O modelo com Kernel polinomial obteve resultados bons e constantes, mesmo alterando o parâmetro C, porém o de melhor desempenho apra este modelo foi o linear.
Para ambos estes modelos, é possível afirmar que para maior mudança entre eles deveria ser variado seu gamma, visto que mudando o parâmetro C não houve mudança.

Para o modelo linear, mesmo não sendo perfeito, ele obteve resultados perto do ideal, sendo um treinamento eficiente para este modelo.
