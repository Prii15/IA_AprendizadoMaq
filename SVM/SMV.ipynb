```python
from sklearn.datasets import load_iris
from sklearn import svm
import matplotlib.pyplot as plt
```

```python
iris_X, iris_y = load_iris(return_X_y = True)

# pega todas as linhas, e apenas coluna 0 e 1
entrada = iris_X[:, :2]
saida = iris_y[:]
```

```python
plt.scatter(entrada[:, 0], entrada[:, 1], c=saida)
plt.xlabel("Comprimento Sépala")
plt.ylabel("Largura Sépala")

# Mostra o gráfico
plt.show()
```

```python
model = svm.SVC(kernel="linear")
model.fit(entrada, saida)
```

```python
model.score(entrada, saida)
```

```python
model = svm.SVC(kernel="rbf", degree=2, gamma='scale', C=5)
model.fit(entrada, saida)
model.score(entrada, saida)
```

```python
entrada = iris_X[:]
saida = iris_y[:]
```

```python
# treinamento com um pedaco da database fornecida
from sklearn.model_selection import train_test_split

entrada_treinamento, entrada_teste, saida_treinamento, saida_teste = train_test_split(entrada, saida, test_size=0.25)

len(entrada_teste)

model = svm.SVC(kernel="poly", degree=7, gamma='auto', C=0.3)
model.fit(entrada_treinamento, saida_treinamento)
```

```python
model.score(entrada_treinamento, saida_treinamento)
```

```python
model.score(entrada_teste, saida_teste)
```

```python
from sklearn.metrics import classification_report

previsto = model.predict(entrada_teste)

print(classification_report(saida_teste, previsto))
```
